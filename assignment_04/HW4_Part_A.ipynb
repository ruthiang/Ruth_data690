## Question #1

```python
with open ("census_costs.txt","r") as f:                  
  line_list = f.readlines()

line_list
```

## Question #2

```python
top2_list = line_list[0:2]
top2_list
```

## Question #3

```python
data_list = line_list[2:]

data_list
```

## Question #4

```python
year_list = []

for line in data_list:
    year = line[0:4]
    year_list.append(year)

print (year_list)
```

## Question #5

```python
pop_list = []

for line in data_list:
    part_list = line.split("\t")
    pop = part_list[1]
    pop_list.append(pop)

print(pop_list)
```

```python
cleansed_pop_list = []

for pop in pop_list:
    cleansed_pop = pop.replace(",","")                  #replacing comma with empty space for pop in the list pop
    cleansed_pop_list.append(int(cleansed_pop))         #adding cleansed_pop to list while making the data type as integers

print(cleansed_pop_list)
```

## Question #6

```python
cost_list = []

for line in data_list:
  part_list = line.split("\t")
  cost = part_list[2]
  cost_list.append(cost)

print(cost_list)
```

```python
cleansed_cost_list = []

for cost in cost_list:
    cleaning_cost = cost.replace(",","")                                  #replacing comma with empty space for cost from cost list
    dollar_cost = cleaning_cost.replace("$","")                           #replacing $ with empty space in cleaning cost and saving it as dollar cost
    if "Billion" in dollar_cost:
        cleansed_cost = int(float(dollar_cost.replace("Billion",""))*1000000000)        # replacing Billions with empty space then changing it into float, multiplying by billion and turning it into integers
    else:
        cleansed_cost = int(dollar_cost)                                  #if no Billion, use dollar cost and turn it into integers

    cleansed_cost_list.append(cleansed_cost)                              #adding cleansed cost variable to cleansed cost list

print(cleansed_cost_list)
```

# Question #7

```python
avg_list = []

for line in data_list:
    part_list = line.split("\t")
    avg = part_list[3]
    avg_list.append(avg)

print(avg_list)
```

```python
cleansed_avg_list = []

for avg in avg_list:
      line_avg = avg.replace("\n","")
      dollar_avg = line_avg.replace("$","")
      if "cents" in dollar_avg:
          cents_avg = float (dollar_avg.replace("cents","")) / 100          #changing cents to empty space, turning it into decimals and multiplying it by 100 to turn it into dollar form
          cents_avg = round(cents_avg, 4)
      else:
          cents_avg = dollar_avg

      cleansed_avg_list.append(cents_avg)

print(cleansed_avg_list)
```

## Question #8 

```python
cleansed_census = list(zip(year_list, cleansed_pop_list, cleansed_cost_list, cleansed_avg_list))

print(cleansed_census)
```

```python
with open("cleansed_census.csv", "w") as w:
   for element in top2_list:                                    #including the top two lines from the original file into the new one
        w.write(element)

   for element in cleansed_census:
        w.write("{0}\t{1}\t{2}\t{3}\n".format(*element))

w.close()
```
